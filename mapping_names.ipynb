```python
# mapping names of folders and file inside them
import os

# The directory containing the files
directory = '/home/somayyeh/Desktop/spicedGithub/repositories/final project/JEPSis/data/MICCAI_BraTS_2018_Data_Training/HGG'

# List all files in the directory
files = os.listdir(directory)

# Filter files to include only those that start with 'Brats18'
brats_files = [f for f in files if f.startswith('Brats18')]

# Sort files if you want the mapping to follow the order in the directory
brats_files.sort()

# Create a mapping from old filenames to new filenames
filename_mapping = {}
for i, old_filename in enumerate(brats_files, start=1):
    new_filename = f'Brats18_{i}'
    filename_mapping[old_filename] = new_filename

# Iterate through each folder in the filename mapping
for old_folder, new_folder_name in filename_mapping.items():
    folder_path = os.path.join(directory, old_folder)
    
    # Ensure the path is a directory before proceeding
    if os.path.isdir(folder_path):
        # List all files in the current directory
        files_in_folder = os.listdir(folder_path)
        
        # Create a new mapping for each file based on its type
        file_mapping = {}
        for file in files_in_folder:
            # Extract the file type (flair, t1, t2, t1ce, seg) from the filename
            if 'flair' in file:
                file_type = 'flair'
            elif 't1ce' in file:
                file_type = 't1ce'
            elif 't1' in file:
                file_type = 't1'
            elif 't2' in file:
                file_type = 't2'
            elif 'seg' in file:
                file_type = 'seg'
            else:
                continue  # Skip if the file type is not recognized

            # Construct the new filename
            new_filename = f'{new_folder_name}_{file_type}.nii'
            file_mapping[file] = new_filename
        
        # Print or use the file mapping as needed
        print(f'Folder: {old_folder} -> {new_folder_name}')
        for old_file, new_file in file_mapping.items():
            print(f'  {old_file} -> {new_file}')

```
